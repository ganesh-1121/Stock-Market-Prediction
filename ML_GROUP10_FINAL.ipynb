## Classifying into clusters

```python
import pandas as pd
from sklearn.cluster import KMeans

# Function to find and print clusters for a dataset
def print_clusters(dataset_name, data, num_clusters):
    # Extract 'Close' prices
    close_prices = data[['Close']]

    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    data['Cluster'] = kmeans.fit_predict(close_prices)

    # Print the cluster assignments
    print(f"Cluster assignments for {dataset_name} Dataset:")
    cluster_table = pd.DataFrame({'Date': data['Date'], 'Cluster': data['Cluster']})
    print(cluster_table)

# Load and process each dataset
# Load the ggl dataset (replace 'path_to_ggl_data.csv' with the actual file path)

print_clusters('ggl', ggl, num_clusters=3)

# Load the inf dataset (replace 'path_to_inf_data.csv' with the actual file path)

inf['Close'] = pd.to_numeric(inf['Close'], errors='coerce')
inf = inf.dropna(subset=['Close'])
print_clusters('inf', inf, num_clusters=3)

# Load the tsl dataset (replace 'path_to_tsl_data.csv' with the actual file path)

print_clusters('tsl', tsl, num_clusters=3)

# Load the f1 dataset (replace 'path_to_f1_data.csv' with the actual file path)

print_clusters('f1', f1, num_clusters=3)
```

## Evalution Metrices

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Function to evaluate clustering models
def evaluate_clustering(data, dataset_name, num_clusters):
    # Extract 'Close' prices
    close_prices = data[['Close']]

    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    data['Cluster'] = kmeans.fit_predict(close_prices)

    # Calculate silhouette score
    silhouette_avg = silhouette_score(close_prices, data['Cluster'])
    print(f"Silhouette Score for {dataset_name} Dataset: {silhouette_avg}")

    # Calculate Davies-Bouldin index
    davies_bouldin_idx = davies_bouldin_score(close_prices, data['Cluster'])
    print(f"Davies-Bouldin Index for {dataset_name} Dataset: {davies_bouldin_idx}")
    
# Load and process each dataset
# Load the ggl dataset (replace 'path_to_ggl_data.csv' with the actual file path)

evaluate_clustering(ggl, 'ggl', num_clusters=3)

# Load the inf dataset (replace 'path_to_inf_data.csv' with the actual file path)

inf['Close'] = pd.to_numeric(inf['Close'], errors='coerce')
inf = inf.dropna(subset=['Close'])
evaluate_clustering(inf, 'inf', num_clusters=3)

# Load the tsl dataset (replace 'path_to_tsl_data.csv' with the actual file path)

evaluate_clustering(tsl, 'tsl', num_clusters=3)

# Load the f1 dataset (replace 'path_to_f1_data.csv' with the actual file path)

evaluate_clustering(f1, 'f1', num_clusters=3)
```

# Loss and Error Curves

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming 'Close' is the column you're clustering on
ggl_close = ggl[['Close']]

# Define a range of K values
K_range = range(1, 11)

# Initialize an empty list to store loss values
loss_values = []

# Iterate through different values of K
for k in K_range:
    # Fit KMeans clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ggl_close)
    
    # Append inertia (loss) values to the list
    loss_values.append(kmeans.inertia_)

# Plot the loss curve (inertia)
plt.figure(figsize=(8, 5))
plt.plot(K_range, loss_values, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Loss)')
plt.title('Loss Curve for Optimal K in KMeans Clustering (ggl Dataset)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Assuming 'Close' is the column you're clustering on
ggl_close = ggl[['Close']]

# Define a range of K values
K_range = range(2, 11)  # Start from K=2 as silhouette score requires at least 2 clusters

# Initialize empty lists to store inertia and silhouette score values
inertia_values = []
silhouette_values = []

# Iterate through different values of K
for k in K_range:
    # Fit KMeans clustering
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ggl_close)
    
    # Append inertia (loss) values to the list
    inertia_values.append(kmeans.inertia_)
    
    # Compute silhouette score and append to the list
    if k > 1:  # Silhouette score calculation requires at least 2 clusters
        silhouette = silhouette_score(ggl_close, kmeans.labels_)
        silhouette_values.append(silhouette)
    else:
        silhouette_values.append(None)  # For K=1, silhouette score is not defined

# Plotting
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plotting inertia values
color = 'tab:blue'
ax1.set_xlabel('Number of Clusters (K)')
ax1.set_ylabel('Inertia (Loss)', color=color)
ax1.plot(K_range, inertia_values, marker='o', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis to plot silhouette score
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Silhouette Score', color=color)
ax2.plot(K_range, silhouette_values, marker='o', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Errors for Optimal K in KMeans Clustering (ggl Dataset)')
fig.tight_layout()
plt.show()
```

# Hyper Parametric Tuning And Evalution Metrics

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score



# Assuming 'Close' and 'Adj Close' are relevant features for clustering
X = ggl[['Close', 'Adj Close']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Grid search for hyperparameter tuning
best_score = -1
best_eps = 0
best_min_samples = 0

for eps in [0.1, 0.5, 1.0, 1.5]:
    for min_samples in [5, 10, 15, 20]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)

        # Check if DBSCAN created more than one cluster
        if len(set(labels)) > 1:
            score = silhouette_score(X_scaled, labels)
            print(f'Parameters: eps={eps}, min_samples={min_samples}, Silhouette Score: {score}')

            if score > best_score:
                best_score = score
                best_eps = eps
                best_min_samples = min_samples

print(f'Best Parameters: eps={best_eps}, min_samples={best_min_samples}, Best Silhouette Score: {best_score}')
```

```python
import pandas as pd
import numpy as np
from hmmlearn import hmm
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score



# Assuming 'Close' is the column of interest
data = np.array(ggl['Close']).reshape(-1, 1)

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Hyperparameter tuning
best_score = -np.inf
best_n_components = None
best_covariance_type = None

# Try different values for n_components and covariance_type
for n_components in range(2, 6):
    for covariance_type in ['spherical', 'tied', 'diag', 'full']:
        # Create an HMM model
        model = hmm.GaussianHMM(n_components=n_components, covariance_type=covariance_type, n_iter=100)

        # Fit the model to the training data
        model.fit(train_data)

        # Predict hidden states for the test data
        hidden_states = model.predict(test_data)

        # Evaluate the model using silhouette score
        silhouette_avg = silhouette_score(test_data, hidden_states)
        print(f"Silhouette Score for n_components={n_components}, covariance_type={covariance_type}: {silhouette_avg}")

        # Update the best parameters if a better silhouette score is found
        if silhouette_avg > best_score:
            best_score = silhouette_avg
            best_n_components = n_components
            best_covariance_type = covariance_type

# Print the best hyperparameters
print(f"Best Hyperparameters - n_components: {best_n_components}, covariance_type: {best_covariance_type}")
```
